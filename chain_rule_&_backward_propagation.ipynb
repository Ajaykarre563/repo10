# **What is mean by chain rule ?**

The chain rule in deep learning helps us calculate how much a change in one part of the model affects the final result, step by step. Since a deep learning model has multiple layers, each depending on the one before it, we use the chain rule to break this process into smaller, manageable steps.

**Think of it like a chain of dominos:**

If you push the first domino, it affects the next one, then the next, and so on.
    
The chain rule tells us how each domino (layer) contributes to the overall effect.

# **what is mean by Backward Propagation ?**

**Backward propagation**, or backpropagation, is the process used in training neural networks to calculate the gradient of the loss function with respect to each model parameter (e.g., weights and biases). These gradients are used to update the parameters and improve the model's predictions.

**The gradient of the loss function** refers to the rate of change of the loss with respect to the model's parameters (such as weights and biases). It tells us how much the loss will increase or decrease if we adjust each parameter slightly.

**Steps in Backpropagation (Simplified)**

**Forward Pass:** The model processes the input to make a prediction, and the loss function calculates the error.
    
**Backward Pass:** Starting from the error, backpropagation calculates how much each layer contributed to it using the chain rule.
    
**Parameter Update:** The model’s weights and biases are adjusted using the gradients to reduce the error in future predictions.

**The multivariate chain** rule helps you find how one thing changes when it depends on several other things. It breaks down the change into smaller steps, looking at how each factor affects the overall change.

**Example: Car Speed and Position**

Imagine you're driving a car, and your position depends on two things:

Speed (how fast you're going), which changes with time.
    
Direction (whether you're going straight or turning), which also changes over time.

Let's say your position is pp, and it's a function of speed (vv) and direction (dd):
p=f(v,d)
p=f(v,d)

**Now, both speed and direction are changing over time. Your speed depends on time, and your direction changes over time too:**

v=g(t)v=g(t)
    
d=h(t)d=h(t)

To find out how your position changes with time, you'd use the multivariate chain rule.
